```python
import re

# -----------------------------
# Phishing Detection Rules
# -----------------------------

phishing_keywords = [
    "urgent",
    "verify",
    "password",
    "bank",
    "account suspended",
    "login",
    "click here",
    "limited time",
    "winner",
    "lottery",
    "free",
    "gift",
    "otp",
    "credit card",
    "dear customer",
    "security alert",
    "update account",
    "confirm identity",
    "verify account",
    "payment failed",
    "claim prize",
    "act now",
    "immediately"
]

# -----------------------------
# Function
# -----------------------------

def analyze_message(message):

    message_lower = message.lower()

    red_flags = []

    score = 0

    # Check keywords

    for word in phishing_keywords:

        if word in message_lower:
            red_flags.append(f"Suspicious keyword detected: {word}")
            score += 1

    # Detect URLs

    urls = re.findall(r'https?://\S+|www\.\S+', message)

    if urls:

        for url in urls:
            red_flags.append(f"Suspicious link: {url}")
            score += 2

            if ".xyz" in url or ".ru" in url:
                red_flags.append("Untrusted domain detected")
                score += 2

    # Email Address

    emails = re.findall(r'\S+@\S+', message)

    for email in emails:

        if "gmail.com" in email or "yahoo.com" in email:
            red_flags.append("Personal email used instead of official company email")
            score += 1

    # Excessive urgency

    urgent_words = ["urgent", "immediately", "now", "today"]

    count = 0

    for word in urgent_words:

        if word in message_lower:
            count += 1

    if count >= 2:
        red_flags.append("Uses excessive urgency")
        score += 2

    # Password Request

    if "password" in message_lower:
        red_flags.append("Requests password")
        score += 2

    # OTP Request

    if "otp" in message_lower:
        red_flags.append("Requests OTP")
        score += 2

    # Bank Details

    if "bank" in message_lower or "credit card" in message_lower:
        red_flags.append("Requests banking information")
        score += 2

    # Risk Level

    if score >= 8:
        level = "HIGH"
        verdict = "PHISHING"

    elif score >= 4:
        level = "MEDIUM"
        verdict = "SUSPICIOUS"

    else:
        level = "LOW"
        verdict = "LIKELY SAFE"

    return verdict, level, red_flags


# Main Program


print("="*60)
print("PHISHING AWARENESS ANALYZER")
print("="*60)

choice = input("1. Type Message\n2. Read from File\nChoose: ")

if choice == "1":

    print("\nPaste your email/message below:\n")
    message = input()

elif choice == "2":

    filename = input("Enter filename: ")

    with open(filename, "r", encoding="utf-8") as file:
        message = file.read()

else:

    print("Invalid Choice")
    exit()

verdict, level, red_flags = analyze_message(message)

print("\n")
print("="*60)
print("ANALYSIS RESULT")
print("="*60)

print("Verdict :", verdict)
print("Risk Level :", level)

print("\nRed Flags Found:")

if red_flags:

    for i, flag in enumerate(red_flags, 1):
        print(f"{i}. {flag}")

else:

    print("No suspicious indicators found.")

print("\nWhy is this unsafe?")

if verdict == "PHISHING":

    print("The message contains multiple phishing indicators including suspicious links, urgent language, and requests for sensitive information.")

elif verdict == "SUSPICIOUS":

    print("Some phishing indicators were detected. Verify the sender before responding.")

else:

    print("No major phishing indicators were detected, but always remain cautious.")
```
